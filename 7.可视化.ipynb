```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig=plt.figure()
# 生成图片，空白绘图窗口
ax1=fig.add_subplot(2,2,1)
# 在fig中创建子图
ax2=fig.add_subplot(2,2,4)
plt.plot(np.random.randn(50).cumsum(),'k--')
# style='k--'黑色分段线
## plt在最后生成的fig或ax上绘制
ax1.scatter(np.arange(10),np.arange(10)*np.random.randn(10))
# 也可直接指定子图
```

```python
# 也可使用fig,axes=plt.subplots(2,3)
# 生成的axes为2行3列，可以是axes[0,1]指定第1行第2列的子图
# 可用plt.subplots_adjust()调节子图wspace，space=0指定子图间间距
fig,axes=plt.subplots(2,2)
axes[0,0].plot(np.random.randn(50).cumsum(),'ko--')
# 等价于color='k',linestyle='dashed',marker='o'
axes[1,1].scatter(np.arange(10),np.arange(10)*np.random.randn(10))
plt.subplots_adjust(wspace=0,hspace=0)
# 无间距会使轴标签重叠
```

```python
fig=plt.figure()
data=np.random.randn(50).cumsum()
plt.plot(data,'k--',label='label1')
plt.plot(data,'k-',drawstyle='steps-post',label='steps-post')
plt.legend(loc='best')
# 图例,有label有legend
```

```python
# plt.xlim([0,1])设置x轴范围，用于截取
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)
plt.plot(np.random.randn(50).cumsum(),'k--')
ticks=ax1.set_xticks([0,10,20,30,40,50])
# 设置x轴刻度
labels=ax1.set_xticklabels(['a','b','c','d','e','f'])
# 刻度标签
ax1.set_title('title')
# 表名称
ax1.set_xlabel('xlabel')
# x轴名称
# 也可以是props={'title':'ti','xlael'='xl'};ax1.set(**props)
```

```python
ds=pd.read_csv('dataset.csv',nrows=20)
ds.head()
```

```python
fig=plt.figure()
ds['Power'].plot()
# Series默认是折线图，索引作x轴
```

```python
ds.plot()
# 可通过subplots=True把每一列放在不同子图
```

```python
ds.plot.bar()
```

```python
ds.plot.barh(stacked=True)
# stacked堆积柱状图
```

```python
import seaborn as sns
```

```python
fig=plt.figure()
# 也需要指定fig或ax
sns.barplot(y='WindSpeed',x='Power',data=ds,orient='h')
# data指明数据来源，x，y列名作为xy轴，orient表示方向
# hue='分隔列名'，如对day（周一，周二）直方图可以按time（早，中，晚）分割
# sns的特点就是传入数据集，然后指明列名
```

```python
fig=plt.figure()
ds["Power"].plot.hist(bins=10)
# hist直方图，频率的离散表示，bins指定箱数
```

```python
fig=plt.figure()
ds["Power"].plot.density()
# 密度图
```

```python
fig=plt.figure()
sns.distplot(ds['Power'],bins=10,color='k')
# 直方图和密度图的融合
```

```python
fig=plt.figure()
sns.regplot("WindSpeed",'Power',ds)
# 画出散点图并拟合出回归曲线
```

```python
# 探索性数据分析中，查看每一组变量的散点图
sns.pairplot(ds,diag_kind='kde',plot_kws={'alpha':0.2})
# diag_kind对角线图类型,plot_kws对非对角线图配置
```

```python
# 分面会图sns.factorplot(x=列名,y,hue，row行分,col按列分,kind图类型,data数据集)
# kind=box分箱图
sns.factorplot(data=ds,x="WindSpeed",y="Power",kind='box')
# 效果不好因为没分箱
```
