```python
ds.to_period('M').head()
# 时间戳转换为区间
```

```python
ds.to_period('M').to_timestamp(how='end').head()
# 区间转换为时间戳，细节丢失 
```

```python
# resample()重新采样，可以看作是groupby的api
ds.resample('1h').mean()
# 对1hgroupby
# 默认是左闭合如11-1，11，3，freq=1d,==11-1,11-2
```

```python
ds.resample('50min',closed='right',label='right',loffset='1s').mean()
# close指定右闭合，label表示索引名用右边,loffset偏置
```

```python
# 金融数据开端-峰值-峰谷-结束（ohlc）
```

```python
# 向上采样，低频率转为高频率
# .resample().ffill(limit=2)对空值前插法，limit指定只对前两个空值填充
```

```python
# 移动窗口函数 
# rolling()算子
ds.rolling(10,min_periods=3).std()
# 滑动窗口大小为10，根据滑动窗口大小分组，这里是10个滑动窗口大小的标准差
# min_periods=3最少观测值数目是3
```

```python
ds.expanding(10).std()
# 扩展窗口
```

```python
ds.rolling('1h').std()
# 这里窗口大小是1h的标准差
# 也可以看作是一种resample或groupby
# rolling().apply()可以自己指定函数
```

```python
# 指数加权函数ewm，给近期的观测值更多权重
ds.ewm(span=10).mean()
```
