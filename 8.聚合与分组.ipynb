```python
import pandas as pd
import numpy as np
```

```python
ds=pd.read_csv('dataset.csv',nrows=20)
ds.tail(5)
```

```python
ds['Map']=ds["WindSpeed"].map(lambda x: int(x))
```

```python
ds["N"]=ds["Map"].map(lambda x:'a' if x<=3 else 'b')
```

```python
ds.head()
```

```python
grouped=ds["WindSpeed"].groupby([ds["N"],ds["Map"]])
# 或ds.group(['N','Map'])['windspeed]
means=grouped.mean()
means
# 根据N，Map分组求windspeed列的统计值
# 使用as_index=False可以使分组键不作为索引
```

```python
means.unstack()
```

```python
ds.groupby(['N']).mean()
```

```python
# groupby()生成二元组，可迭代
# for (k1,k2),group in df.groupby(['M'],'N')
# 或dict（list（df.groupby()））生成字典
# 可以用axis指定轴
```

```python
# 也可传入数组
dict1={'Map':'M','N':'N','Power':'O',"RotorSpeed":"O"}
# 两个O合并，对列进行
cc=ds.groupby(dict1,axis=1)
cc.sum()
```

```python
# 对多级索引进行分组，可以指定level
```

```python
# 数据聚合，可以分好再用，或者使用df.groupby().agg(f)使用自己定义的聚合函数
grouped.describe()
```

```python
# agg中可以传递list，指定多个函数
grouped.agg(['mean',('元组第一个字符作名称','std')])
```

```python
# agg({'lie1':'fun1','lie2':'fun2'})传入字典对不同列使用不同函数,agg聚合
```

```python
grouped.agg(['mean','sum'])
```

```python
ds.groupby('N').apply(lambda ds:ds.sort_values(by='Power'))
## 根据分组，对每个分组进行应用apply，每个分组分别调用函数
# 函数被各分组调用，然户concat在一起
```

```python
ds.groupby('N',group_keys=False).apply(lambda ds:ds.sort_values(by='Power'))
# 取消分组的索引
```

```python
cutts=(pd.qcut(ds.WindSpeed,4,labels=False))
# 分箱也可作为groupby
```

```python
group1=ds.groupby(cutts).sum()
group1
```

```python
# 数据透视表和交叉表
ds.pivot_table(['Power',"WindSpeed"],index=['Map','N'])
```

```python
ds.pivot_table(['Power',"WindSpeed"],index=['Map'],columns="N",margins=True)
# 根据行m列n透视拍p,ws,margin添加行列标签和all
# 还可指定聚合函数aggfunc=，填充空值fill_value=
```

```python
pd.crosstab(ds.Map,ds.N,margins=True)
# 统计分组中的频率，行，列
```
