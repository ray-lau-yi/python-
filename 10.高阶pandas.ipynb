```python
import pandas as pd
import numpy as np
```

```python
ds=pd.read_csv('dataset.csv',nrows=20)
```

```python
ds["cat"]=['a','b','c','d']*5
```

```python
ds.head()
```

```python
# 可以用astypej转为categroy
ds.cat=ds.cat.astype('category')
ds.cat
# 也可直接用pd.Categorical
```

```python
ds.cat.values
# 值是Categories实例，具有categories和codes属性
```

```python
ds.cat.values.categories
```

```python
ds.cat.values.codes
```

```python
# 可以依靠codes和categories生成Catgory
cat1=pd.Categorical.from_codes(ds.cat.values.codes,ds.cat.values.categories,ordered=True)
# 指定顺序，因此对于已有cat可以用cat.as_ordered()排序
cat1
```

```python
# Category与groupby搭配性能更好
# %time ... ...的时间开销
```

```python
# 特殊属性cat提供了对分类方法的访问
ds['cat'].cat.codes
```

```python
ds.cat.cat.categories
```

```python
ds.cat.cat.remove_unused_categories()
# 去除未观察到的分类，数量为0
```

```python
# 生成独热码
pd.get_dummies(ds.cat).head()
```

```python
# groupby
ds.groupby('cat').mean()
```

```python
ds.groupby('cat').agg('mean')
```

```python
ds.groupby('cat').transform('mean')
# 每行被替代，尺寸和原来相同
```
